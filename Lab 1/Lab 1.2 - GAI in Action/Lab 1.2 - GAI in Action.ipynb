#<em><strong>Exercise 1: Book Recommendations</strong></em>

In this hands-on exercise, you'll explore how to use ChatGPT to generate personalized book recommendations for different age groups, genres, and interests by crafting effective prompts. This activity will help you understand the power of prompting and how AI can assist in supporting readers and library users.



### Step 1: Open ChatGPT in a New Tab and Sign In

[ChatGPT](https://chatgpt.com/)



### Step 2: Give ChatGPT some Background Info

Click on your profile in the top right of the screen. Then click "Customize ChatGPT." This allows you to enter information into the GPT about you in order for it to give you the best output. Additionally you can also customize how you want ChatGPT to respond (professionally, conversationally, like a gen alpha, etc.)

GPT1.png


For this activity, make up a character and tell the GPT their age, favorite genres, books they have read and liked (few-shot prompting!), and books they did not like in the "Anything else ChatGPT to know about you?" section.




### Step 3: Test it out!

Now in a  New Chat, ask it for some book recommendations! Because you already specified your preferences in the "Customize ChatGPT" screen, you should not have to specify again what this character's favorite books are.


Play around with different book preferences and see what ChatGPT recommends!

Examples:


```
Teenager who likes mystery novels
Graphic novels about coming-of-age stories
Children's books with diverse characters
```

# <em><strong>Exercise 3: Creating Library Flyers and Newsletters Using Generative AI</strong></em>


In this activity, you will learn how to use ChatGPT and generative tools to create promotional content such as flyers and newsletters for the library. Youll practice writing prompts to generate event flyers, weekly newsletters, and corresponding images using AI.


### Step 1: Start a New Chat

Create a new chat by clicking the button in the top left.

---
## **Part A: Writing a Weekly Library Newsletter**

### Step 2: Use a Prompt to Create a Newsletter

Copy the prompt into the new chat:


> You are a librarian at a community college. Write a friendly, engaging weekly newsletter for students. Include:
- A new book or featured collection
- An upcoming event or workshop
- A study tip of the week
- Library hours for the weekend
Keep it under 200 words and use a casual tone.


### Step 3: Review and Customize

Ask ChatGPT to rewrite or adjust the tone, layout, or content if needed.

Optional: Ask it to format the newsletter for email or print.

---

## **Part B: Creating a Flyer for a Library Event**

### Step 4: Prompt ChatGPT to Write a Flyer Text


Copy this prompt:

> Create a one-page flyer text for an upcoming Lakeside Library event:<br>
"Poetry Night Open Mic  Thursday at 6 PM in the Reading Lounge. Free snacks provided. Bring your own poem or read a favorite."
Make it eye-catching, informative, and fun.





### Step 5: Ask or alternative versions


You can try variations for different audiences:
>Can you rewrite the flyer for social media use?<br>
Can you add a short quote about poetry to make it more inspiring?

---




## **Part C: Generating Images to Match the Flyer or Newsletter**

### Step 6: Generate a Visual Using DALL-E (available in ChatGPT)

Use this prompt:


> Create an image of a cozy college library reading lounge with students performing poetry in the evening. Include warm lighting and bookshelves.



### Step 7: Review and Download the Image

After generation, click to enlarge and download the image if you like it.
You can also ask ChatGPT to try a different style (e.g., watercolor, digital art, realistic).

---

## **Part D: Writing a Library Blog Post**

### Step 8: Prompt ChatGPT to Write a Blog Post

Copy this prompt:

> Write a 300-word blog post for the library blog introducing new graphic novels added this month. Mention at least three titles and describe why students might enjoy them. Write in an engaging, informative tone.



### Step 9: Adjust for Tone, Length, or Audience

You can ask:
>Can you rewrite this post for younger readers?

or

>Add a call-to-action encouraging students to visit the library and check out the books.

## **Reflection Questions**
1.   What types of content did AI help you create most effectively?
2. How can blog posts support student engagement with library resources?
3. What other communications could AI help the library team generate?
